```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
#pandas - here we are using excel 
```

```python
final_df = pd.merge(df,df_country,on = 'Country Code', how='left')

# we are merging two datasets
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
#stores all country names
```

```python
country_values = final_df.Country.value_counts().values
#values for all countries
```

```python
#Pie Chart - Top 3 countries
plt.rcParams['figure.figsize']=(12,8)
plt.pie(country_values[:3],labels = country_names[:3],autopct='%1.2f%%')
```

Observation: most transcations are from india then US and then UK 

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()

#change the index by using reset_index
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})

#rename(columns={0:'Rating Count'})  - used to change the column name
```

```python
ratings
```

## observations 
1.when rating is 4.5 to 4.9 - excellent
2.when rating is 4.4 to 4.0 - verygood
3.when rating is 3.9 to 3.4 - good
4.when rating is 3.4 to 2.5 - average
5.when rating is 2.4 to 1.8 - poor

```python
plt.figure(figsize=(12,7))
sns.barplot(x='Aggregate rating',y='Rating Count', data = ratings)
```

```python
plt.rcParams['figure.figsize']=(12,8)

sns.barplot(x='Aggregate rating',y='Rating Count',hue = 'Rating color', data = ratings)

#hue is used to seperate with legend
```

```python
plt.rcParams['figure.figsize']=(12,8)

sns.barplot(x='Aggregate rating',y='Rating Count',hue = 'Rating color', data = ratings,palette = ['blue', 'red', 'orange', 'yellow', 'green','green'])
```

Observations:
    1. NOT rated is very high
    2.maximum num of ratings are 2.5 to 3.4 

```python
## count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue', 'red', 'orange', 'yellow', 'green','green'])

#count here is represents frequency of rating
```
